```python
%matplotlib inline  
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('ANFR_Herault_Mesures.csv')
df.head()
```

```python
df.rename(columns={'Niveau global': 'Niveau_global'}, inplace=True)
```

```python
df=df.assign(target=1) #création d'une nouvelle colonne
print(df)
```

### On a crée une nouvelle colonne "Target" . Par défaut, toute la colonne est égale à 1. Alors, j'ai 912 lignes et 17 colonnes.

```python
df.columns # cette ligne permet d'obtenir les différentes colonnes que j'ai dans Dataset.
```

```python
def target():
    for t in range(0,len(df)):
        if df['Niveau_global'][t] > 6 : 
            df['target'][t] = 0
            print(df['Niveau_global'][t]) 
```

### La fonction target permet de vérifier si le Niveau_global > 6 alors Target=0 

```python
target()
```

```python
pd.crosstab(df.target, "freq") 
```

### J'obtient :  14 valeurs supérieur à 6 et 898 valeurs inférieurs à 6 . (seuil = 6)

```python
df.head(30) # Les 30 lignes premiers 
```

```python
X=df[['Numéro de mesure','Niveau_global']].to_numpy() # x a 2 colonnes : Numéro de mesure et Niveau_global
print(X)
y=df['target'].to_numpy() # y = est un vecteur contient Target
print(y)
```

### ==> Ma Dataset n'est pas équilibré car on a plus des 1 que les 0 . On 14 zéros 

### Maintenant , on va faire l'entrainement de mon modéle  

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0) 
```

### - On a 30% pour le teste et 70% pour le train 

```python
from sklearn.linear_model import LogisticRegression 
#Create a l.Reg Classifier
logreg1 = LogisticRegression()
#Train the model using the training sets
logreg1.fit(X_train, y_train)
y_pred=logreg1.predict(X_test)
```

### Le modéle de machine learning que nous allons utiliser est 'LOGISTIC REGRESSION' 

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

### LOGISTIC REGRESSION peut donner jusqu'à 80% de précision , on peut dire que c'est un bon modéle.
